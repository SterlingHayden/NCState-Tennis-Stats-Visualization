```python
import pandas as pd
```

```python
df_excel = pd.read_excel('Data/Final2023-24NCAAIWomensTennisWTAPointsSinceNCAAsMay262024.xlsx')
df_excel.head()
```

## Top 10 singles players

```python
columns_points = ['Singles', 'Singles.1', 'Singles.2', 'Singles.3', 'Singles.4', 'Singles.5', 'Singles.6', 'Singles.7']
columns_players = ['Player 1', 'Player 2', 'Player 3', 'Player 4', 'Player 5', 'Player 6', 'Player 7', 'Player 8']

# Flatten the data across the selected columns
all_values = df_excel[columns_points].values.flatten()
all_players = df_excel[columns_players].values.flatten()

# We will flatten the 'Thru 5/26/2024' column to match the shape of the points and players
all_schools = df_excel['Thru 5/26/2024'].repeat(len(columns_points)).values

# Combine points, players, and school names
points_players_schools = list(zip(all_values, all_players, all_schools))

# Sort by points and get the top 10 values with corresponding players and school names
top_10_points_players_schools = sorted(points_players_schools, key=lambda x: x[0], reverse=True)[:10]

# Convert to DataFrame
df_top_10 = pd.DataFrame(top_10_points_players_schools, columns=['Singles Points', 'Player', 'School'])

# Display the DataFrame
df_top_10
```

## Top 10 doubles players


```python
columns_points = ['Doubles', 'Doubles.1', 'Doubles.2', 'Doubles.3', 'Doubles.4', 'Doubles.5', 'Doubles.6', 'Doubles.7']
columns_players = ['Player 1', 'Player 2', 'Player 3', 'Player 4', 'Player 5', 'Player 6', 'Player 7', 'Player 8']

# Flatten the data across the selected columns
all_values = df_excel[columns_points].values.flatten()
all_players = df_excel[columns_players].values.flatten()

# We will flatten the 'Thru 5/26/2024' column to match the shape of the points and players
all_schools = df_excel['Thru 5/26/2024'].repeat(len(columns_points)).values

# Combine points, players, and school names
points_players_schools = list(zip(all_values, all_players, all_schools))

# Sort by points and get the top 10 values with corresponding players and school names
top_10_points_players_schools = sorted(points_players_schools, key=lambda x: x[0], reverse=True)[:10]

# Convert to DataFrame
df_top_10 = pd.DataFrame(top_10_points_players_schools, columns=['Doubles Points', 'Player', 'School'])

# Display the DataFrame
df_top_10
```
